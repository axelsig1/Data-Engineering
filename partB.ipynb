```python
from pyspark.sql import SparkSession
from operator import add

spark_session = SparkSession.builder\
        .master("spark://192.168.2.250:7077") \
        .appName("Assignment-3-Axel-Sigstam")\
        .config("spark.dynamicAllocation.enabled", True)\
        .config("spark.dynamicAllocation.shuffleTracking.enabled",True)\
        .config("spark.shuffle.service.enabled", False)\
        .config("spark.dynamicAllocation.executorIdleTimeout","30s")\
        .config("spark.executor.cores", 2)\
        .config("spark.driver.port",9999)\
        .config("spark.blockManager.port",10005)\
        .getOrCreate()

# RDD API
spark_context = spark_session.sparkContext

spark_context.setLogLevel("INFO")
```

```python
########
## B1 ##
########

# read the file from the HDFS
parking_df = spark_session.read.csv('hdfs://192.168.2.250:9000/parking-citations.csv')
parking_df.show()
```

```python
########
## B2 ##
########

# Print the schema
parking_df.printSchema()
```

```python
########
## B3 ##
########

# Count the number of rows
row_count = parking_df.count()
print("Number of rows in the CSV file:", row_count)
```

```python
# Number of rows in the CSV file: 13077725
```

```python
########
## B4 ##
########

# Count the number of partitions in the underlying RDD
num_part = parking_df.rdd.getNumPartitions()
print("Number of partitions in the underlying RDD:", num_part)
```

```python
# Number of partitions in the underlying RDD: 16
```

```python
########
## B5 ##
########

# Drop the columns VIN, Latitude and Longitude.
parking_df = parking_df.drop("VIN", "Latitude", "Longitude")
parking_df.show()
```

```python
########
## B6 ##
########
from pyspark.sql.functions import col

# From above we see that "fine amount" is column name "_c16"

# Convert 'fine amount' column to float
parking_df = parking_df.withColumn("_c16", col("_c16").cast("float"))

# Find the maximum amount
max_fine_amount = parking_df.agg({"_c16": "max"}).collect()[0][0]

# Count the number of fines with the maximum amount
num_max_fines = parking_df.filter(col("_c16") == max_fine_amount).count()

print("Maximum Fine Amount:", max_fine_amount)
print("Number of Fines with Maximum Amount:", num_max_fines)
```

```python
# Maximum Fine Amount: 1100.0
# Number of Fines with Maximum Amount: 626
```

```python
########
## B7 ##
########
from pyspark.sql.functions import desc

# Group by 'Vehicle Make' and count occurrences
make_counts = parking_df.groupBy("_c8").count()

# Sort in descending order based on count
make_counts_sorted = make_counts.orderBy(desc("count"))

# Show the top 20 most frequent vehicle makes
make_counts_sorted.show(20)
```

```python
########
## B8 ##
########

from pyspark.sql.functions import udf
from pyspark.sql.types import StringType

# Define the dictionary of colors
COLORS = {
    'AL': 'Aluminum', 'AM': 'Amber', 'BG': 'Beige', 'BK': 'Black',
    'BL': 'Blue', 'BN': 'Brown', 'BR': 'Brown', 'BZ': 'Bronze',
    'CH': 'Charcoal', 'DK': 'Dark', 'GD': 'Gold', 'GO': 'Gold',
    'GN': 'Green', 'GY': 'Gray', 'GT': 'Granite', 'IV': 'Ivory',
    'LT': 'Light', 'OL': 'Olive', 'OR': 'Orange', 'MR': 'Maroon',
    'PK': 'Pink', 'RD': 'Red', 'RE': 'Red', 'SI': 'Silver', 'SL': 'Silver',
    'SM': 'Smoke', 'TN': 'Tan', 'VT': 'Violet', 'WT': 'White', 'WH': 'White',
    'YL': 'Yellow', 'YE': 'Yellow', 'UN': 'Unknown'
}

# Function to map abbreviations to their full names
def expand_color(color):
    return COLORS.get(color, color)

# Register the UDF
expand_color_udf = udf(expand_color, StringType())

# Apply the UDF to create the new column 'color_long'
# Take color from column "_c10"
parking_df = parking_df.withColumn('color_long', expand_color_udf(parking_df['_c10']))

# Show the DataFrame with the new column
parking_df.show()
```

```python
# Select only the 'color_long' column and show it
parking_df.select('color_long').show()
```

```python
########
## B9 ##
########
from pyspark.sql.functions import col

# Filter the DataFrame for Toyotas and group by the 'color_long' column, then count the occurrences
toyota_colors = parking_df.filter(col('_c8') == 'TOYT') \
                          .groupBy('color_long') \
                          .count() \
                          .orderBy(col('count').desc())

# Show the most frequent color value for Toyotas
most_frequent_color_for_toyotas = toyota_colors.first()
print("Most frequent color value for Toyotas:", most_frequent_color_for_toyotas['color_long'])
```

```python
# Most frequent color value for Toyotas: Gray
# Seems plausible as gray is one of the most frequent car colors over all.
```
